**Audio translator with transcriptions**

---


This audio translator translates a given audio in different indian languages and provides audio and transcription for the same .

>Input :Any audio with language support in -
*Afrikaans, Arabic, Armenian, Azerbaijani, Belarusian, Bosnian, Bulgarian, Catalan, Chinese, Croatian, Czech, Danish, Dutch, English, Estonian, Finnish, French, Galician, German, Greek, Hebrew, Hindi, Hungarian, Icelandic, Indonesian, Italian, Japanese, Kannada, Kazakh, Korean, Latvian, Lithuanian, Macedonian, Malay, Marathi, Maori, Nepali, Norwegian, Persian, Polish, Portuguese, Romanian, Russian, Serbian, Slovak, Slovenian, Spanish, Swahili, Swedish, Tagalog, Tamil, Thai, Turkish, Ukrainian, Urdu, Vietnamese, and Welsh.*
and size<25 MB



>Output :

* Text transcription in -Bangla,Gujrati,Hindi, Kannada,Malyalam,Tamil,Telegu,Punjabi,Sindhi,Urdu
* Audio translation in -Bangla,Gujrati,Hindi, Kannada,Malyalam,Tamil,Telegu,Urdu

> The translator has an **automatic language detection system** which identifies the language of the given audio on it's own


> The translator also provides accuracy score of translation by word error rate










Importing Dependenices

```python
!pip -qqq install git+https://github.com/openai/whisper.git
```

```python
import whisper
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
whisper_model = whisper.load_model("large", device=device)
```

Speech to text

```python
audio_file = '/content/audio.mp3'
result = whisper_model.transcribe(audio_file)
```

```python
print(result["text"])
```

Detecting the language of transcribed text

```python
!pip install langdetect
```

```python
from langdetect import detect

# Detect the language code
language_code = detect(result['text'])

print(f"Detected language code: {language_code}")
```

Translation in different indian languages

```python
!pip install googletrans==4.0.0-rc1
```

```python
from googletrans import Translator
from googletrans import LANGUAGES
```

```python
translations = {}
indian_language_codes = ['bn', 'gu', 'hi', 'kn', 'ml', 'pa', 'sd', 'ta', 'te', 'ur']
translator = Translator()
# Iterate through the Indian language codes and translate the text
for lang_code in indian_language_codes:
    try:
        translation = translator.translate(text=result["text"], src=language_code, dest=lang_code)
        translated_text = translation.text
        print(f"Translation in {lang_code}: {translated_text}")
        translations[lang_code] = translated_text
    except Exception as e:
        print(f"Error translating to {lang_code}: {str(e)}")
```

Calculating WER

```python
from nltk.metrics.distance import edit_distance
```

```python
wer_scores={}
for lang_code, translated_text in translations.items():
    # Tokenize the reference and translated texts into words
    reference_words = result['text'].split()
    translated_words = translated_text.split()

    # Calculate WER
    wer = edit_distance(reference_words, translated_words)
    wer_rate = wer / len(reference_words)

    # Store the WER for the translation
    wer_scores[lang_code] = wer_rate

print(wer_scores)
```

Returning the translated audio in different languages

```python
!pip install gTTS
```

```python
from gtts import gTTS
import os

output_dir = 'translated_audio'

# Create the output directory
os.makedirs(output_dir, exist_ok=True)
for lang_code, translated_text in translations.items():
    try:
        tts = gTTS(text=translated_text, lang=lang_code)
        audio_file_path = os.path.join(output_dir, f'{lang_code}_translation.mp3')
        tts.save(audio_file_path)
        print(f'Audio generated for {lang_code}')
    except ValueError as e:
        print(f"Language not supported for {lang_code}: {str(e)}")
```
